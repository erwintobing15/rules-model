```python
import pandas as pd
import matplotlib as plt

# read dataset
dataset = pd.read_csv('clean_data.csv')
del dataset['Unnamed: 0']
dataset.head()
```

```python
# data information
dataset.info()
```

```python
# class distribusion
dataset['radiant_win'].value_counts()
```

### Distribusi dan sebaran pemilihan hero untuk id hero 1, 20, 40, 60, 80, 100, 120, 135
Nilai -1 menunjukan id hero dipilih oleh tim dire, 0 tidak dipilih oleh kedua tim, dan 1 menunjukkan id hero dipilih oleh tim radiant

```python
import seaborn as sns

# variable distribution and dispersion
id_sample = ['1', '20', '40', '60', '80', '100', '120', '135']
dataset[id_sample].hist(bins=15, figsize=(15, 6), layout=(2, 4));
```

### Sebaran label

```python
sns.countplot(dataset['radiant_win']);
```

### Relasi antara fitur dengan fitur dan dengan label (2 dimensi)

```python
# id hero 1 with 8
sns.jointplot(x=dataset['1'], y=dataset['8']);
```

```python
# id hero 1 with label radiant_win
sns.jointplot(x=dataset['1'], y=dataset['radiant_win']);
```

### Representasi relasi antara hero 1 dan 2 terhadap label radiant_win (3 dimensi)

```python
sns.relplot(data=dataset, x="1", y="2", hue="radiant_win")
```

### Representasi relasi antara hero 1, 2, 3, 4 terhadap label radiant_win (5 dimensi)

```python
sns.lmplot(x="1", y="2",
           hue="radiant_win", col="3", row="4",
           height=3, data=dataset)
```
